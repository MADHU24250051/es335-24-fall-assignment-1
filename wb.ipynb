```python
# Function to prepare the data for training and testing
def prepare_data_for_subject(test_subject):
    # Initialize lists to store data and labels
    all_features = []
    all_labels = []

    for activity in activities:
        for subject in subjects:
            if subject == test_subject:
                continue  # Skip the test subject

            file_path = os.path.join(csv_directory, f'{activity}{subject}.csv')
            data = pd.read_csv(file_path)

            # Extract 2 samples from the selected range if enough data is available
            if len(data) > 801:  # Ensure at least 801 rows for sampling
                features = data.loc[800:801, ['ax (m/s^2)', 'ay (m/s^2)', 'az (m/s^2)']]  # Select rows 800 to 801
                labels = pd.Series([activity] * len(features))

                # Normalize features
                scaler = StandardScaler()
                features_scaled = scaler.fit_transform(features)

                all_features.append(features_scaled)
                all_labels.append(labels)

    # Concatenate all features and labels into single arrays
    X = np.vstack(all_features)
    y = np.concatenate(all_labels)

    # Ensure balanced samples for each activity and subject in the training set
    activity_samples = defaultdict(list)
    for feature, label in zip(X, y):
        activity_samples[label].append(feature)

    # Collect exactly 2 samples for each activity per subject
    num_samples_per_activity =  2* (len(subjects) - 1)  # 2 samples for each of the remaining 2 subjects
    X_balanced = []
    y_balanced = []

    for activity in activities:
        samples = activity_samples[activity][:num_samples_per_activity]
        X_balanced.extend(samples)
        y_balanced.extend([activity] * len(samples))

    # Convert lists to numpy arrays
    X_balanced = np.array(X_balanced)
    y_balanced = np.array(y_balanced)

    # Prepare test data for the left-out subject
    test_samples = []
    test_labels = []

    if test_subject:
        for activity in activities:
            file_path = os.path.join(csv_directory, f'{activity}{test_subject}.csv')
            data = pd.read_csv(file_path)
            
            # Extract samples for testing
            if len(data) > 805:  # Ensure at least 806 rows for sampling
                features = data.loc[800:801, ['ax (m/s^2)', 'ay (m/s^2)', 'az (m/s^2)']]  # Select rows 800 to 801
                labels = pd.Series([activity] * len(features))

                # Normalize features
                scaler = StandardScaler()
                features_scaled = scaler.fit_transform(features)

                test_samples.append(features_scaled)
                test_labels.append(labels)

    # Concatenate test samples and labels
    X_test = np.vstack(test_samples)
    y_test = np.concatenate(test_labels)

    return X_balanced, X_test, y_balanced, y_test
```

```python
# Function to map model output to standard label
def map_model_output_to_label(output):
    output = output.lower()
    if 'walking upstairs' in output:
        return 'walking_upstairs'
    elif 'walking downstairs' in output:
        return 'walking_downstairs'
    elif 'walking' in output:
        return 'walking'
    elif 'sitting' in output:
        return 'sitting'
    elif 'standing' in output:
        return 'standing'
    else:
        return 'laying'
```

### Zero-shot

```python
# Function to classify test samples and calculate accuracy
def classify_and_evaluate(X_test, y_test, llm):
    true_labels = []
    predictions = []

    # Loop through each test sample and classify it using the model
    for i in range(len(X_test)):
        test_sample = X_test[i]
        true_label = y_test[i]
        
        query = f"""
        You are a human activity recognition model. Based on the following accelerometer data, classify the activity:
        Data: {test_sample.tolist()}
        Classify the activity as 'walking', 'walking_upstairs', 'walking_downstairs', 'sitting', 'standing', 'laying'.
        """
        
        # Invoke the model
        try:
            answer = llm.invoke(query)
            predicted_label = map_model_output_to_label(answer.content)
        except Exception as e:
            print(f"Error: {e}")
            predicted_label = 'unknown'  # Fallback label

        # Store the prediction and true label
        predictions.append(predicted_label)
        true_labels.append(true_label)

    # Calculate accuracy
    accuracy = accuracy_score(true_labels, predictions)

    # Print the true labels and corresponding predictions for all test samples
    for i in range(len(true_labels)):
        print(f"Sample {i+1}:")
        print(f"True Label: {true_labels[i]}")
        print(f"Predicted Label: {predictions[i]}")
        print()

    return accuracy,true_labels,predictions
```

### Few-shot

```python
# Function to create Few-Shot Learning prompt
def create_few_shot_prompt(examples, test_sample):
    example_text = "\n".join(
        [f"Example {i+1} - Data: ax={example['data'][0]}, ay={example['data'][1]}, az={example['data'][2]}, Activity: {example['activity']}"
            for i, example in enumerate(examples)
        ]
    )
    
    return f"""
    You are a human activity recognition model. Based on the following accelerometer data, classify the activity.
    Here are some examples for reference:
    {example_text}
    
    Now classify the following data:
    Data: ax={test_sample[0]}, ay={test_sample[1]}, az={test_sample[2]}
    Classify the activity as 'walking', 'walking_upstairs', 'walking_downstairs', 'sitting', 'standing', 'laying'.
    """

# Function to select Few-Shot examples
def select_few_shot_examples(X, y, num_samples_per_column=2):
    examples = []
    indices = random.sample(range(len(X)), num_samples_per_column)
    for index in indices:
        examples.append({'data': X[index], 'activity': y[index]})
    return examples
```

```python
acc=[]
# Function to classify test samples and calculate accuracy
def classify_samples(X_test, y_test, llm):
    fs_true_labels = []
    fsl_predictions = []

    # Loop through each test sample and classify it using the Few-Shot prompt
    for i in range(len(X_test)):
        test_sample = X_test[i]
        true_label = y_test[i]
        
        # Create Few-Shot Learning prompt
        few_shot_query = create_few_shot_prompt(few_shot_examples, test_sample)
        
        # Invoke the model
        answer = llm.invoke(few_shot_query)
        
        # Map the model output to an activity label
        predicted_label = map_model_output_to_label(answer.content)
        
        # Store the prediction and true label
        fsl_predictions.append(predicted_label)
        fs_true_labels.append(true_label)
        

    # Calculate and print accuracy
    correct_predictions = sum(true == pred for true, pred in zip(fs_true_labels, fsl_predictions))
    accuracy = correct_predictions / len(fs_true_labels) * 100
    acc.append(accuracy)
    print(f"Accuracy: {accuracy:.2f}%")

    # Print the true labels and corresponding predictions for all test samples
    for i in range(len(fsl_predictions)):
        print(f"Sample {i+1}:")
        print(f"True Label: {fs_true_labels[i]}")
        print(f"Predicted Label: {fsl_predictions[i]}")
        print()
    return acc,fs_true_labels,fsl_predictions
    
    # Loop over each subject, leaving them out one at a time
for test_subject in subjects:
    print(f"\nProcessing with Subject {test_subject} as the test subject...\n")
    
    X_train, X_test, y_train, y_test = prepare_data_for_subject(test_subject)
    
    # Print the sizes of the datasets
    print(f"Training set size: {len(X_train)}")
    print(f"Test set size: {len(X_test)}")
    
    # Select Few-Shot examples from the training data
    few_shot_examples = select_few_shot_examples(X_train, y_train)
    
    # Classify the samples using the Few-Shot Learning model
    fsl_acc,fs_true_labels, fs_pred = classify_samples(X_test, y_test, llm)
```
