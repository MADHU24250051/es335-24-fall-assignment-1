### Zero shot Promting

```python
# Function to map model output to standard label
def map_model_output_to_label(output):
    output = output.lower()
    if 'walking upstairs' in output:
        return 'walking_upstairs'
    elif 'walking downstairs' in output:
        return 'walking_downstairs'
    elif 'walking' in output:
        return 'walking'
    elif 'sitting' in output:
        return 'sitting'
    elif 'standing' in output:
        return 'standing'
    else:
        return 'laying'
```

```python
import numpy as np
import pandas as pd

acc_data_sample1 = pd.read_csv('/home/hp/Desktop/ML/es335-24-fall-assignment-1/HAR/Combined/Train/LAYING/Subject_1.csv')
acc_data_sample2 = pd.read_csv('/home/hp/Desktop/ML/es335-24-fall-assignment-1/HAR/Combined/Train/SITTING/Subject_11.csv')
acc_data_sample3 = pd.read_csv('/home/hp/Desktop/ML/es335-24-fall-assignment-1/HAR/Combined/Train/WALKING/Subject_7.csv')

example1 = np.array(acc_data_sample1)
example2 = np.array(acc_data_sample2)
example3 = np.array(acc_data_sample3)

examples = [example1, example2, example3]
true_labels = ["laying", "sitting", "walking"]

predictions = []

for i, example in enumerate(examples):
    test_data = example[500:502]  
    query = f"""
    * You are a human activity recognition model. 
    * Your task is to classify the following accelerometer data as Walking, Walking_upstairs, Walking_downstairs, Sitting, Standing, Laying.
    * Output should contain predicted label and reason

    Sentence: {test_data}
    """ 

    answer = llm.invoke(query)
    prediction = map_model_output_to_label(answer.content)
    predictions.append(prediction)
    
    print(f"True label: {true_labels[i]}, Predicted label: {prediction}")


correct_predictions = sum([pred == true for pred, true in zip(predictions, true_labels)])
accuracy = correct_predictions / len(true_labels)

print(f"Accuracy: {accuracy * 100:.2f}%")


    
```

### Few shot Promting

```python

predictions = []

# few-shot examples
few_shot_examples = """
Example 1:
Input: [0.6918975,-0.1267594,0.2531432]
Label: Sitting
Example 2:
Input: [0.9263129,0.1705382,0.3739298]
Label: Sitting
Example 3:
Input: [0.1827699,0.5370223,0.8485901]
Label: Laying
Example 4:
Input: [0.1874844,0.7878636,0.5772027]
Label: Laying
Example 5:
Input: [1.00865,-0.2501793,0.07970832]
Label: Standing
Example 6:
Input: [0.9995704,-0.264085,0.08752285]
Label: Standing
Example 7:
Input: [1.531087,-0.5873186,-0.3011638]
Label: Walking
Example 8:
Input: [1.401607,-0.3784688,0.1549877]
Label: Walking
Example 9:
Input: [1.680843,-0.5298075,0.01739741]
Label: Walking_upstairs
Example 10:
Input: [0.7287424,-0.2633399,0.1466003]
Label: Walking_upstairs
Example 11:
Input: [1.034825,-0.3044757,0.06453713]
Label: Walking_downstairs
Example 12:
Input: [0.5573862,-0.08842983,-0.01346502]
Label: Walking_downstairs



"""

for i, example in enumerate(examples):
    test_data = example[500:505]  

    # Prepare the query with few-shot examples
    query = f"""
    * You are a human activity recognition model. 
    * Your task is to classify the following accelerometer data as Walking, Walking_upstairs, Walking_downstairs, Sitting, Standing, Laying.
    * Output should contain predicted label and reason.
    
    Few-shot examples:
    {few_shot_examples}

    Sentence: {test_data}
    """ 

    answer = llm.invoke(query)
    prediction = map_model_output_to_label(answer.content) 
    predictions.append(prediction)
    
    print(f"True label: {true_labels[i]}, Predicted label: {prediction}")


correct_predictions = sum([pred == true for pred, true in zip(predictions, true_labels)])
accuracy = correct_predictions / len(true_labels)

print(f"Accuracy: {accuracy * 100:.2f}%")
```

1) In the above accuracies of both zero shot and few shot learning methods, the performance of both methods is poor.

```python
# Function to create few-shot examples from training data
def create_few_shot_examples(train_folder, activities, num_examples_per_activity=3):
    few_shot_examples = ""
    for activity in activities:
        file_path = os.path.join(train_folder, f"{activity}.csv")
        
        if os.path.exists(file_path):
            data = pd.read_csv(file_path).loc[:, ['accx', 'accy', 'accz']].values  
            
            # Load a few examples from the CSV file
            for i in range(min(num_examples_per_activity, len(data))):  
                example_data = data[i]  
                example_str = ", ".join(map(str, example_data))  
                
                few_shot_examples += f"Example {i + 1} ({activity}):\nInput: [{example_str}]\nLabel: {activity}\n\n"
        else:
            print(f"File {file_path} does not exist.")
    return few_shot_examples

# Generate few-shot examples from training data
few_shot_examples = create_few_shot_examples(train_folder, activities)
```

```python
# Function to map model output to standard label
def map_model_output_to_label(output):
    output = output.lower()
    if 'walking upstairs' in output:
        return 'walking_upstairs'
    elif 'walking downstairs' in output:
        return 'walking_downstairs'
    elif 'walking' in output:
        return 'walking'
    elif 'sitting' in output:
        return 'sitting'
    elif 'standing' in output:
        return 'standing'
    else:
        return 'laying'
```

```python
# Function to predict using the model and calculate accuracy
def predict_and_calculate_accuracy(test_folder, activities, few_shot_examples):
    predictions = []
    true_labels = []

    for activity in activities:
        file_path = os.path.join(test_folder, f"{activity}.csv")
        
        if os.path.exists(file_path):
            data = pd.read_csv(file_path).loc[:, ['ax (m/s^2)', 'ay (m/s^2)', 'az (m/s^2)']].values  

            # Select exactly 2 samples from test data for prediction
            for i in range(2):  
                if i < len(data):
                    test_data = data[i] 

                    # Convert test data to a string format for the prompt
                    test_data_str = ", ".join(map(str, test_data)) 

                    # Create the prompt for the model
                    query = f"""
                    * You are a human activity recognition model.
                    * Your task is to classify the following accelerometer data as Walking, Walking_upstairs, Walking_downstairs, Sitting, Standing, Laying.
                    * Output should contain predicted label and reason.
                    
                    Few-shot examples:
                    {few_shot_examples}
                    
                    Sentence: [{test_data_str}]
                    """

                    answer = llm.invoke(query)
                    prediction = map_model_output_to_label(answer.content)

                    predictions.append(prediction)
                    true_labels.append(activity)
                    
                    print(f"True label: {activity}, Predicted label: {prediction}")
        else:
            print(f"File {file_path} does not exist.")

    # Calculate accuracy
    correct_predictions = sum([pred == true for pred, true in zip(predictions, true_labels)])
    accuracy = correct_predictions / len(true_labels)
    
    print(f"Accuracy: {accuracy * 100:.2f}%")

# prediction and accuracy calculation
predict_and_calculate_accuracy(test_folder, activities, few_shot_examples)
```
